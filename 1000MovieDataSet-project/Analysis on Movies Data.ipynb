# **8. Exploratory Data Analysis**
---

- This section is emphasised on asking the right questions and perform analysis using the data.

- Note that there is no limit how deep you can go, but make sure not to get distracted from right track.

```python
# Now that we have cleaned the dataset, we can proceed with visualizing the data to gain insights. Let's start by looking at the distribution of the movie ratings:
# Plotting the distribution of movie ratings
sns.histplot(df['Rating'])
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.show()
```

### From the plot, we can see that the majority of movies have a rating between 60 and 80. There are also a few movies with a rating above 9.

```python
# Next, let's explore the relationship between the movie ratings and the revenue generated by the movies:

# Plotting the relationship between movie ratings and revenue
sns.scatterplot(x=df['Rating'], y=df['Revenue (Millions)'])
plt.title('Relationship between Movie Ratings and Revenue (Millions)')
plt.xlabel('Rating')
plt.ylabel('Revenue (Millions)')
plt.show()
```

### From the plot, we can see that there is a positive correlation between the movie ratings and revenue generated. This means that movies with higher ratings tend to generate more revenue.

```python
#Let's also explore the relationship between the movie ratings and the number of votes:

# Plotting the relationship between movie ratings and votes
sns.scatterplot(x=df['Rating'], y=df['Votes'])
plt.title('Relationship between Movie Ratings and Votes')
plt.xlabel('Rating')
plt.ylabel('Votes')
plt.show()
```

### From the plot, we can see that there is a positive correlation between the movie ratings and the number of votes. This means that movies with higher ratings tend to have more votes.

```python
#Now, let's explore the distribution of movie genres:

# Splitting the genres and creating a list of all genres
genres_list = []
for genres in df['Genre']:
    genres_list.extend(genres.split(','))

# Creating a dataframe of all genres and their frequencies
genres_df = pd.DataFrame(genres_list, columns=['Genre'])
genres_df['Count'] = 1
genres_count = genres_df.groupby('Genre')['Count'].sum().reset_index().sort_values(by='Count', ascending=False)

# Plotting the distribution of movie genres
sns.barplot(x='Genre', y='Count', data=genres_count)
plt.title('Distribution of Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

### From the plot, we can see that the most popular movie genres are Drama, Action, Comedy, and Thriller.

```python
#let's explore the relationship between the movie runtime and revenue:

# Plotting the relationship between movie runtime and revenue
sns.scatterplot(x=df['Runtime (Minutes)'], y=df['Revenue (Millions)'])
plt.title('Relationship between Movie Runtime and Revenue')
plt.xlabel('Runtime')
plt.ylabel('Revenue')
plt.show()
```

### From the plot, we can see that there is a positive correlation between the movie runtime and revenue generated. This means that movies with longer runtimes tend to generate more revenue

```python
#Let's explore the most successful directors in the dataset based on the revenue generated by their movies. We will create a barplot of the top 20 directors and their total revenue

# Creating a dataframe of total revenue by director
director_revenue = df.groupby('Director')['Revenue (Millions)'].sum().reset_index().sort_values(by='Revenue (Millions)', ascending=False)

# Plotting the top 20 directors by revenue
sns.barplot(x='Revenue (Millions)', y='Director', data=director_revenue.head(20))
plt.title('Top 20 Directors by Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Director')
plt.show()
```

### From the plot, we can see that J.J. Abrams is the most successful director in the dataset in terms of revenue generated by his movies

```python
# Creating a list of all actors and their revenue
actors_revenue = []
for i, row in df.iterrows():
    for actor in row['Actors'].split(','):
        actors_revenue.append((actor.strip(), row['Revenue (Millions)']))

# Creating a dataframe of total revenue by actor
actors_revenue_df = pd.DataFrame(actors_revenue, columns=['Actor', 'Revenue (Millions)'])
actor_revenue = actors_revenue_df.groupby('Actor')['Revenue (Millions)'].sum().reset_index().sort_values(by='Revenue (Millions)', ascending=False)

# Plotting the top 20 actors by revenue
sns.barplot(x='Revenue (Millions)', y='Actor', data=actor_revenue.head(20))
plt.title('Top 20 Actors by Total Revenue')
plt.xlabel('Total Revenue (in billions)')
plt.ylabel('Actor')
plt.show()
```

### From the plot, we can see that Robert Downey Jr. is the most successful actor in the dataset in terms of revenue generated by the movies he starred in.
